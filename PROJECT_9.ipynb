```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

while True:
    print("Please select an option : ")
    print("1. Load Dataset")
    print("2. Explore Data")
    print("3. Perform DataFrame Operation")
    print("4. Handle Missing Data")
    print("5. Generate Descriptive Statistics")
    print("6. Data Visualization")
    print("7. Save Visualization ")
    print("8. Exit")
    print("")
    c = int(input("Enter the Choice -> "))
    if c == 1:
        try:
            path = input("Enter the path of csv file -> ")
            df = pd.read_csv(path)
            print("Dataset Loaded Successfully!\n")
        except Exception as e:
            print("Error: Unable to load dataset. Please check the path again.")
            print("Details:", e)

    elif c == 2:
        while True:
            print("=== Explore Data ==")
            print("1. Display the first 5 rows")
            print("2. Display the last 5 rows")
            print("3. Display column names")
            print("4. Display data types")
            print("5. Display basic info")
            print("6. Exit")
            i = int(input("Enter your Choice -> "))
            if i == 1:
                print("First 5 Rows\n")
                print(df.head(5))
            elif i == 2:
                print("Last 5 Rows\n")
                print(df.tail(5))
            elif i == 3:
                print("Display Columns\n")
                co = df.columns
                for i in co:
                    print(i)
            elif i == 4:
                print("Data Types\n")
                print(df.dtypes)
            elif i == 5:
                print("Basic Info Of Data Set\n")
                print(df.info())
            elif i == 6:
                break

            else:
                print("Invalid Choice")

    elif c == 3:
        while True:
            print(" Perform DataFrame Operation")
            print("1. Check the how much Data Is Null")
            print("2. Describe Dataset")
            print("3. Exit")
            m = int(input("Enter the "))
            if m == 1:
                print("Null Data Count ")
                null = df.isnull().sum()
                print(null)
            elif m == 2:
                print("Describe Dataset \n")
                print(df.describe())
            elif m == 3:
                break
            else:
                print("Invalid Choice")

    elif c == 4 :
        while True:
            print(" Handle Missing Data ")
            print("1. Display rows with missing values")
            print("2. Fill missing values with mean")
            print("3. Drop rows with missing values")
            print("4. Replace missing values with a specific ")
            print("5. Exit")
            h = int(input("Enter the choice"))
            if h == 1:
                print("Rows with missing values\n", df.isnull())
            elif h == 2:
                print("Fill missing value with mean\n")
                fill = df.fillna().mean()
                print(fill)
            elif h == 3:
                df.dropna()
                print("Dataset Drop Null Value Successfully")
            elif h == 4:
                df.fillna("NAN")
            elif h == 5:
                break
            else:
                print("Invalid Choice")

    elif c == 5:
        print("Generate Descriptive Statistics")
        print(df.describe())

    elif c == 6:
        while True:
            print("Data Visualization")
            print("1. Bar Plot")
            print("2. Line Plot")
            print("3. Scatter Plot")
            print("4. Pie Plot")
            print("5. Histogram")
            print("6. Stack Plot")
            print("7. Exit")
            v = int(input("Enter your Choice"))
            x = input("Enter x-axis Column name")
            y = input("Enter y-axis column name")
            if v == 1:
                plt.figure(figsize=(8, 7))
                plt.bar(df[x], df[y])
                plt.title("Bar Plot")
                plt.show()
            elif v == 2:
                plt.figure(figsize=(8, 7))
                plt.plot(df[x],df[y], marker='o')
                plt.title("Line Plot")
                plt.show()
            elif v == 3:
                plt.figure(figsize=(8, 7))
                plt.scatter(df[x],df[y])
                plt.title("Line Plot")
                plt.show()
            elif v == 4:
                plt.figure(figsize=(8, 7))
                plt.pie(df[y], labels=df[x], autopct='%1.1f%%')
                plt.title("Pie Chart")
                plt.show()
            elif v == 5:
                plt.figure(figsize=(8, 7))
                sns.histplot(data=df,x=x)
                plt.title("Histogram Plot")
                plt.show()
            elif v == 6:
                plt.figure(figsize=(8, 7))
                plt.stackplot(df[x], df[y])
                plt.title("Stack Plotting")
                plt.show()
            elif v == 7:
                print("Thanks")
                break
            else:
                print("Invalid Choice")
    elif c == 7:
        save = input("Enter File Name To save Plot")
        plt.savefig(save)
        plt.show()

    elif c == 8:
        print("Thank You.")
        break
    else:
        print("Invalid Choice!!try again..")
```
